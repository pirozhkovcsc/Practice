# Train dataset

Здесь мы будем предсказывать значения синуса.

Сначала создадим тренировочный датасет.

```python
x_train = torch.rand(100)
x_train = x_train * 20.0 - 10.0 # 100 рандомных точек от -10 до 10

y_train = torch.sin(x_train)

plt.plot(x_train.numpy(), y_train.numpy(), 'o')
plt.title('$y = sin(x)$');
```

```python
noise = torch.randn(y_train.shape) / 5. 
# Это шум, который представляет собой 100 чисел из стандартного нормального распределения, которые поделили на 5

# Теперь график train выборки и noise (в последствии мы добавим этот шум к y_train)
plt.plot(x_train.numpy(), noise.numpy(), 'o')
plt.axis([-10, 10, -1, 1])
plt.title('Gaussian noise');
```

```python
y_train = y_train + noise # Добавили шум

# Теперь посмотрим, как выглядит зашумленный синус
plt.plot(x_train.numpy(), y_train.numpy(), 'o')
plt.title('noisy sin(x)')
plt.xlabel('x_train')
plt.ylabel('y_train');
```

```python
x_train.unsqueeze_(1) # Здесь мы превратили x_train в вектор столбец
y_train.unsqueeze_(1); # Здесь мы превратили y_train в вектор столбец
```

```python
# В качестве примера, как работает метод unsqueeze
tmp = torch.Tensor([1,2,3])
print(tmp)
print(tmp.unsqueeze(1))
```

# Validation dataset

Создадим теперь датасет, на котором будем валидироваться.

```python
x_validation = torch.linspace(-10, 10, 100)
y_validation = torch.sin(x_validation.data)

plt.plot(x_validation.numpy(), y_validation.numpy(), 'o')
plt.title('sin(x)')
plt.xlabel('x_validation')
plt.ylabel('y_validation');
```

```python
x_validation.unsqueeze_(1)
y_validation.unsqueeze_(1);
```

# Model construction

Здесь мы создадим архитектуру.

```python
class SineNet(torch.nn.Module):
    def __init__(self, n_hidden_neurons):
        super(SineNet, self).__init__()
        self.fc1 = torch.nn.Linear(1, n_hidden_neurons)
        self.act1 = torch.nn.Sigmoid()
        self.fc2 = torch.nn.Linear(n_hidden_neurons, 1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.fc2(x)
        return x

sine_net = SineNet(100)
```

# Prediction

Попробуем не обучаясь сразу предсказать синус.

```python
def predict(net, x, y):
    y_pred = net.forward(x)

    plt.plot(x.numpy(), y.numpy(), 'o', label='Groud truth')
    plt.plot(x.numpy(), y_pred.data.numpy(), 'o', c='r', label='Prediction');
    plt.legend(loc='upper left')
    plt.xlabel('$x$')
    plt.ylabel('$y$')

predict(sine_net, x_validation, y_validation)
```

# Optimizer

Поставим оптимизатор, который будем регулировать наши параметры в сети. lr ставим равным 0.01.

```python
optimizer = torch.optim.Adam(sine_net.parameters(), lr=0.01)
```

# Loss function

Определим функцию потерь. Здесь это MSE.

```python
def loss(pred, target):
    squares = (pred - target) ** 2
    return squares.mean()
```

# Training procedure

Самая важная часть - это тренировка данных.

Каждый последовательный шаг обучения называется эпохой. Т.е подкрутили один раз параметры = прошли одну эпоху.

Пройдем 2000 эпох. Итак, сначала у оптимизатора сбросим градиент (можно было бы это делать и в конце), затем предскажем значения y_pred с помощью нашей сети и тренировочной выборки x_train. Далее вычислим значение функции потерь. После посчитаем градиент функции потерь и сделаем итерационный шаг (как задумано в Adam). Одна такая итерация обновляет нам параметры. Т.е по сути, мы градиентным спуском приходим к наилучшим, в смысле функции потерь, параметрам. Проделав 2000 итераций - получаем хорошие параметры сети (веса, смещения).

Теперь предскажем с этими параметрами значения и нарисуем график:)

```python
for epoch_index in range(2000):
    optimizer.zero_grad()

    y_pred = sine_net.forward(x_train)
    loss_val = loss(y_pred, y_train)

    loss_val.backward()

    optimizer.step()

predict(sine_net, x_validation, y_validation)
```
