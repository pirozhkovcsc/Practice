## Задача 1.

Здесь будет реализован метод наискорейшего спуска. Данную задачу можно посмотреть [здесь](https://drive.google.com/file/d/1q3XLkKtb8e6DBn_5X1EbbFHuSAXN0j6k/view), она находится под номером 1.

```python
import numpy as np
```

Введем размерность:

```python
n = int(input()) # Размерность
```

Теперь создадим функцию, которая будет возвращать нам трехдиагональную матрицу:

```python
from scipy.sparse import diags

def tri_diag(a, b, c):
    return diags(shape=(n, n), diagonals=[a, b, c], offsets=[-1, 0, 1]).toarray()
```

С помощью функции выше создадим матрицу T и D:

```python
T = tri_diag(-0.25, 1, -0.25) # Matrix T
D = tri_diag(0, -0.25, 0) # Matrix D
```

Теперь создадим функцию, которая возвращает блочную трехдиагональную матрицу:

```python
from scipy.linalg import block_diag

def block_tri_diag(A, B, C):
    temp_1 = block_diag(*[B]*n)
    temp_2 = np.concatenate([block_diag(*[A]*n)[n:], np.zeros(shape=(n, n ** 2))])
    temp_3 = np.concatenate([np.zeros(shape=(n, n ** 2)), block_diag(*[C]*n)[:(n ** 2 - n)]])
    return temp_1 + temp_2 + temp_3
```

Наконец, создадим блочную трехдиагональную матрицу как в нашем условии (на диагоналях $D, T, D$):

```python
Mat = block_tri_diag(D, T, D) # Head matrix
```

Когда мы создали матрицу, осталось реализовать градиентный спуск. Подробнее про него можно прочитать [здесь](http://samarskii.ru/books/book1989.pdf), страница 119.

```python
def GD(Mat, x_prev, b, stop=0.001):
    counter = 0
    while True:
        r = Mat @ x_prev - b
        if np.sqrt(r.T @ r) < stop:
            break
        tau = (r.T @ r) / ((Mat @ r).T @ r)
        x_new = x_prev - tau * r
        x_prev = x_new
        counter += 1
    return x_prev, counter
```

```python
solve, iteration = GD(Mat, x_prev=np.zeros(n ** 2).reshape(-1, 1), b=np.ones(n ** 2).reshape(-1, 1), stop=0.000001)
print(f'Приближенное решение:\n {solve}')
print(f'Количество итераций: {iteration}')
```
