```python
import pandas as pd
import numpy as np

from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_score, recall_score, plot_roc_curve, plot_confusion_matrix, \
    plot_precision_recall_curve, accuracy_score

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize': (9, 6)})
```

```python
train = pd.read_csv('https://stepik.org/media/attachments/course/4852/train_iris.csv', index_col=0)
```

```python
train.head()
```

```python
X_train = train.drop(['species'], axis=1)
y_train = train.species
```

```python
train.head()
```

```python
test = pd.read_csv('https://stepik.org/media/attachments/course/4852/test_iris.csv', index_col=0)
```

```python
test.head()
```

```python
X_test = test.drop(['species'], axis=1)
y_test = test.species
np.random.seed(0)
```

```python
from sklearn.metrics import accuracy_score

scores = pd.DataFrame()
for max_depth in range(1, 100):
    my_tree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    my_tree.fit(X_train, y_train)
    train_score = my_tree.score(X_train, y_train)
    test_score = my_tree.score(X_test, y_test)
    
    temp = pd.DataFrame({'max_depth': [max_depth], 'train_score': [train_score], 'test_score': [test_score]})
    scores = scores.append(temp)
scores.head()
```

```python
scores = pd.melt(
    scores,
    id_vars=['max_depth'],
    value_vars=['train_score', 'test_score'],
    var_name='type',
    value_name='score'
)
scores
```

```python
sns.lineplot(x='max_depth', y='score', hue='type', data=scores);
```
